```python
from scipy.io import arff
import pandas as pd
import glob
import json

import os

DATASET_COLL_DIR = "/home/karma/Documents/AutoML Project/automml-2022-pro/datasets"
DATASET_NAMES = ["enron", "genbase", "image", "langLog", "reuters", "scene", "slashdot", "yeast", "birds", "emotions"]

pd.set_option("display.max.columns", None)
pd.set_option("display.precision", 2)
```

```python
# Read the each dataset from the collection along with its json file

datasets = {}

for name in DATASET_NAMES:
    dataset_dir_path = os.path.join(DATASET_COLL_DIR, name)
    [arff_file_path] = glob.glob(f"{dataset_dir_path}/*.arff")
    [json_file_path] = glob.glob(f"{dataset_dir_path}/*.json")
    
    data, meta = arff.loadarff(arff_file_path)
    df = pd.DataFrame(data)
    
    info = json.load(open(json_file_path))
    desc = info["data_set_description"]["description"]
    target_attrs = info["data_set_description"]["default_target_attribute"].split(",")
    
    datasets.update({name : {"description" : desc, "df" : df, "target_attrs": target_attrs}})
```

## Dataset statistics and Target attributes

```python
def get_details(name):
    df = datasets[name]["df"]
    desc = datasets[name]["description"]
    target_attrs = datasets[name]["target_attrs"]

    print(f"{name.capitalize()}\n")
    print(f"{desc}\n")
    df.info()
    print(f"Number of target attributes is {len(target_attrs)}\n")
    print(f"Target attribute set : \n {target_attrs}")
    df.describe(include="all")
```
